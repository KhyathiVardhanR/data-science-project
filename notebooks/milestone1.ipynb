```python
from pathlib import Path
import pandas as pd

RAW_PATH = Path("../data/raw/TradeData_2_14_2026_20_28_56.csv")
PROCESSED_PATH = Path("../data/processed/pharma_imports.csv")

# load file (your working settings)
df = pd.read_csv(
    RAW_PATH,
    encoding="latin1",
    sep=None,
    engine="python",
    index_col=False
)

# keep only required columns
df_clean = df[["reporterISO", "refYear", "primaryValue"]]

# rename to project schema
df_clean = df_clean.rename(columns={
    "reporterISO": "country",
    "refYear": "year",
    "primaryValue": "pharma_imports"
})

# ensure one row per country-year
df_clean = (
    df_clean.groupby(["country","year"], as_index=False)
            .sum()
)

# sort
df_clean = df_clean.sort_values(["country","year"])

# save
PROCESSED_PATH.parent.mkdir(parents=True, exist_ok=True)
df_clean.to_csv(PROCESSED_PATH, index=False)

print("Saved to:", PROCESSED_PATH)
display(df_clean.head())
```

```python
import pandas as pd
from pathlib import Path

RAW = Path("../data/raw")
OUT = Path("../data/processed/trade_dataset.csv")

# ---------- helper function ----------
def load_trade_files(prefix, value_name):
    files = sorted(RAW.glob(f"{prefix}_*.csv"))
    dfs = []

    for f in files:
        df = pd.read_csv(
            f,
            encoding="latin1",
            sep=None,
            engine="python",
            index_col=False
        )

        df = df[["reporterISO","refYear","primaryValue"]]
        df = df.rename(columns={
            "reporterISO":"country",
            "refYear":"year",
            "primaryValue": value_name
        })

        dfs.append(df)

    df_all = pd.concat(dfs, ignore_index=True)

    # aggregate to country-year
    df_all = (
        df_all.groupby(["country","year"], as_index=False)
              .sum()
    )

    return df_all


# ---------- load each dataset ----------
pharma = load_trade_files("pharma", "pharma_imports")
fuel   = load_trade_files("fuel",   "fuel_imports")
oil    = load_trade_files("oil",    "oil_exports")

# ---------- merge all ----------
trade = pharma.merge(fuel, on=["country","year"], how="outer")
trade = trade.merge(oil, on=["country","year"], how="outer")

# ---------- sort ----------
trade = trade.sort_values(["country","year"])

# ---------- save ----------
OUT.parent.mkdir(parents=True, exist_ok=True)
trade.to_csv(OUT, index=False)

print("Saved final trade dataset to:", OUT)
display(trade.head())
print("Shape:", trade.shape)
```

```python
from pathlib import Path
import pandas as pd

RAW = Path("../data/raw")
ucdp_file = RAW / "UcdpPrioConflict_v25_1.csv"

df = pd.read_csv(ucdp_file, encoding="latin1")

# --- keep only needed columns ---
df = df[["gwno_loc", "year", "intensity_level"]]

# remove rows without country code
df = df.dropna(subset=["gwno_loc"])

# some rows have multiple countries like "651, 666"
df["gwno_loc"] = df["gwno_loc"].astype(str)
df = df.assign(gwno_loc=df["gwno_loc"].str.split(",")).explode("gwno_loc")

# convert to numeric
df["gwno_loc"] = df["gwno_loc"].str.strip().astype(float)

# --- create conflict incidence ---
df["conflict_incidence"] = (df["intensity_level"] > 0).astype(int)

# --- create simple intensity measure ---
df["conflict_intensity"] = df["intensity_level"]

# --- keep only project years ---
df = df[(df["year"] >= 1995) & (df["year"] <= 2025)]

# --- aggregate to country-year ---
conflict = (
    df.groupby(["gwno_loc","year"], as_index=False)
      .agg({
          "conflict_incidence":"max",
          "conflict_intensity":"max"
      })
)

conflict.head()
```

```python
import pycountry

def gw_to_iso(gw):
    try:
        return pycountry.countries.get(numeric=str(int(gw)).zfill(3)).alpha_3
    except:
        return None

conflict["country"] = conflict["gwno_loc"].apply(gw_to_iso)

conflict = conflict.dropna(subset=["country"])

conflict = conflict[["country","year","conflict_incidence","conflict_intensity"]]

conflict.head()
```

```python
from pathlib import Path
import pandas as pd
import pycountry

RAW = Path("../data/raw")
polity_file = RAW / "p5v2018.xls"

# reload file fresh
df = pd.read_excel(polity_file)

# keep only needed columns
df = df[["country","year","polity2"]]

# rename variable
df = df.rename(columns={"polity2":"regime_score"})

# keep only project years
df = df[(df["year"] >= 1995) & (df["year"] <= 2025)]

# convert country name -> ISO3
def name_to_iso(name):
    try:
        return pycountry.countries.lookup(name).alpha_3
    except:
        return None

df["country"] = df["country"].apply(name_to_iso)

df = df.dropna(subset=["country"])

polity = df[["country","year","regime_score"]]

# save
OUT = Path("../data/processed/polity_dataset.csv")
OUT.parent.mkdir(parents=True, exist_ok=True)
polity.to_csv(OUT, index=False)

print("Saved to:", OUT)
polity.head()
```

```python
import pandas as pd
from pathlib import Path

PROC = Path("../data/processed")
RAW  = Path("../data/raw")

trade = pd.read_csv(PROC / "trade_dataset.csv")
political = pd.read_csv(PROC / "political_dataset.csv")

gdp = pd.read_csv(RAW / "gdp_growth.csv", skiprows=4)
infl = pd.read_csv(RAW / "inflation.csv", skiprows=4)
unemp = pd.read_csv(RAW / "unemployment.csv", skiprows=4)

print("Loaded all datasets")
```

```python
def reshape_wdi(df, value_name):
    df = df.melt(id_vars=["Country Code"], var_name="year", value_name=value_name)
    df = df.rename(columns={"Country Code":"country"})
    df["year"] = pd.to_numeric(df["year"], errors="coerce")
    df = df[(df["year"] >= 1995) & (df["year"] <= 2025)]
    return df.dropna(subset=[value_name])
```

```python
gdp_clean   = reshape_wdi(gdp, "gdp_growth")
infl_clean  = reshape_wdi(infl, "inflation")
unemp_clean = reshape_wdi(unemp, "unemployment")

gdp_clean.head()
```

```python
macro = gdp_clean.merge(infl_clean, on=["country","year"], how="outer")
macro = macro.merge(unemp_clean, on=["country","year"], how="outer")

macro.head()
print("Macro shape:", macro.shape)
```

```python
final = trade.merge(political, on=["country","year"], how="outer")
final = final.merge(macro, on=["country","year"], how="outer")

final = final.sort_values(["country","year"])

print("Final preview:")
display(final.head())

print("Final shape:", final.shape)
```

```python
import pandas as pd
from pathlib import Path

PROC = Path("../data/processed")
RAW = Path("../data/raw")

# ---- Load trade & political ----
trade = pd.read_csv(PROC / "trade_dataset.csv")
political = pd.read_csv(PROC / "political_dataset.csv")

# ---- Reload World Bank files ----
gdp = pd.read_csv(RAW / "gdp_growth.csv", skiprows=4)
infl = pd.read_csv(RAW / "inflation.csv", skiprows=4)
unemp = pd.read_csv(RAW / "unemployment.csv", skiprows=4)

# ---- reshape function ----
def reshape_wdi(df, value_name):
    df = df.melt(id_vars=["Country Code"], var_name="year", value_name=value_name)
    df = df.rename(columns={"Country Code":"country"})
    df["year"] = pd.to_numeric(df["year"], errors="coerce")
    df = df[(df["year"] >= 1995) & (df["year"] <= 2025)]
    return df.dropna(subset=[value_name])

# ---- build macro again ----
gdp_clean   = reshape_wdi(gdp, "gdp_growth")
infl_clean  = reshape_wdi(infl, "inflation")
unemp_clean = reshape_wdi(unemp, "unemployment")

macro = gdp_clean.merge(infl_clean, on=["country","year"], how="outer")
macro = macro.merge(unemp_clean, on=["country","year"], how="outer")

# ---- create final dataset ----
final = trade.merge(political, on=["country","year"], how="outer")
final = final.merge(macro, on=["country","year"], how="outer")
final = final.sort_values(["country","year"])

# ---- SAVE final dataset ----
final.to_csv(PROC / "final_dataset.csv", index=False)

print("Final dataset saved.")
```
