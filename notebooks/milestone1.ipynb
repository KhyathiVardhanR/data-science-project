## Data Acquisition: Economic Indicators (World Bank)

This section loads annual economic indicator data from the World Bank,
including inflation, unemployment, and GDP growth. These indicators
represent key measures of civilian economic well-being.


```python
import pandas as pd

inflation = pd.read_csv(
    "data/raw/inflation.csv",
    skiprows=4
)

inflation.head()
```

```python
unemployment = pd.read_csv("data/raw/unemployment.csv", skiprows=4)
unemployment.head()
                        
```

```python
gdp = pd.read_csv("data/raw/gdp_growth.csv", skiprows=4)
gdp.head()
```

```python
inflation.shape
inflation.isna().sum().head()
inflation.describe()
```

```python
unemployment.shape
unemployment.isna().sum().head()
unemployment.describe()
```

```python
gdp.shape
gdp.isna().sum().head()
gdp.describe()
```
